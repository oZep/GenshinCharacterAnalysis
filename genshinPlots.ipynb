```python
#June 21, 2022 - August 10, 2022
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
#opening the csv file
file = open("/Users/joeyissa/Downloads/Genshin_Impact_All_Character_Stat.csv")
#making a file reader
csvReader = csv.reader(file)
#skip first line
next(csvReader)
#convert data into list
data = []
for row in csvReader:
    data.append(row)


#create the empty list
dataChar = []
#you got to compare the list to something at first
prev = ""
#keep track of all the same indexes within the list
lastIndex = -1

for row in data:
    if row[0] != prev:
        #make a new list, wihtin the bigger list, since the name has changed
        dataChar.append([])
        #when time increases you know a new name appeared, at this index
        lastIndex += 1
        #this resets the cycle
        prev = row[0]
    #find the last index (where the new name appeared)
    #append that index (row) to the new list created in the for loop
    dataChar[lastIndex].append(row)

print(dataChar)
```

```python
#find and print certain information from each inner link

#gather information about Atk increases
def Atk(name):

    alistforAtk = []
    n = -1
    lastIndex = -1

    #for each row in the data set
    for row in dataChar:
        #break it apart
        #for each entry in the index row
        for entry in row:
            #within the row, print the seperate entries
            if entry[0] == name:
                #print(entry[8])
                #add the entry to a list
                atk = entry[8]
                lastIndex += 1
                alistforAtk.append([])
                #make the new comparison that entry
                alistforAtk[lastIndex].append(int(atk))
    #print(alistforAtk)
    return (alistforAtk)


    
```

```python
#find and print certain information from each inner link
#gather information about Level increases

def Level(name):
    alistforLevel = []
    n = -1
    lastIndex = -1

    #for each row in the data set
    for row in dataChar:
        #break it apart
        #for each entry in the index row
        for entry in row:
            #within the row, print the seperate entries
            if entry[0] == name:
                #print(entry[1])
                #add the entry to a list
                level = entry[1]
                lastIndex += 1
                alistforLevel.append([])
                #make the new comparison that entry
                alistforLevel[lastIndex].append(int(level))
    #print(alistforLevel)
    return (alistforLevel)

```

```python
#find and print certain information from each inner link
#gather information about HP increases

def hp(name):
    alistforHP = []
    n = -1
    lastIndex = -1

    #for each row in the data set
    for row in dataChar:
        #break it apart
        #for each entry in the index row
        for entry in row:
            #within the row, print the seperate entries
            if entry[0] == name:
                #print(entry[7])
                #add the entry to a list
                HP = entry[7]
                lastIndex += 1
                alistforHP.append([])
                #make the new comparison that entry
                alistforHP[lastIndex].append(int(HP))
    #print(alistforHP)
    return (alistforHP)
```

```python
#find and print certain information from each inner link

def defs(name):
    
    alistforDEF = []
    n = -1
    lastIndex = -1

    #for each row in the data set
    for row in dataChar:
        #break it apart
        #for each entry in the index row
        for entry in row:
            #within the row, print the seperate entries
            if entry[0] == name:
                #print(entry[9])
                #add the entry to a list
                DEF = entry[9]
                lastIndex += 1
                alistforDEF.append([])
                #make the new comparison that entry
                alistforDEF[lastIndex].append(int(DEF))
    #print(alistforDEF)
    return (alistforDEF)
```

```python
#Plot Atk vs Level
def atkplot(name, alistforLevel, alistforAtk):
    plt.title(name)
    #plt.plot(x,y)
    plt.plot( alistforLevel , alistforAtk )
    plt.xlabel("Character Level")
    plt.ylabel("Character Atk")
    plt.show()


```

```python
#Plot HP vs Level
def hpplot(name, alistforLevel, alistforHP):
    plt.title(name)
    #plt.plot(x,y)
    plt.plot( alistforLevel , alistforHP )
    plt.xlabel("Character Level")
    plt.ylabel("Character HP")
    plt.show()
```

```python
#plot Def vs Level
def defplot(name, alistforLevel, alistforDEF):
    plt.title(name)
    #plt.plot(x,y)
    plt.plot(alistforLevel , alistforDEF)
    plt.xlabel("Character Level")
    plt.ylabel("Character DEF")
    plt.show()
```

```python
#find and print certain information from each inner link
#gather information about Character Names

alistforNames = []
#Clock
n = -1
lastIndex = -1
#comparison
priName = " "

#for each row in the data set
for row in dataChar:
    #break it apart
    #for each entry in the index row
    for entry in row:
        #within the row, print the seperate entries
        if entry[0] != priName:
            #print(entry[0])
            #add the entry to a list
            priName = entry[0]
            lastIndex += 1
            alistforNames.append([])
            #make the new comparison that entry
            alistforNames[lastIndex].append(str(priName))
#print(alistforNames)

#-------------

#go through all the names
for row in alistforNames:
    for entry in row:
        name = entry
        #call all the functions
        Atk(name)
        Level(name)
        hp(name)  
        defs(name)
        atkplot(name, alistforLevel, alistforAtk)
        hpplot(name, alistforLevel, alistforHP)
        defplot(name, alistforLevel, alistforDEF)
    
```
