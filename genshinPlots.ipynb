```python
x = np.linspace(0, 3*np.pi, 500)
plt.plot(x, np.sin(x+5), x, x, x, np.sin(x+5)*-1, x, -x)
plt.arrow(2, 0, 5, 0, length_includes_head=True, head_width=1)
plt.title('A simple chirp')
plt.show()
```
